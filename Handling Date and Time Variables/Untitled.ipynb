# working with Dates

```python
# Converting to datatime datatype 
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

# 1. Extract year

```python
date['date_year'] = date['date'].dt.year
date.head()
```

# 2. Extract month

```python
date['date_month'] = date['date'].dt.month
date.head()
```

# 3. Extract days

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# here we extrcting the month from the data 
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

# Extract the day of week

```python
date['date_dow'] = date['date'].dt.dayofweek
date.head()  # here we exrtring the date of week from the date 
```

```python
# day of week - name 

date['date_dow_name'] = date['date'].dt.day_name()   # here the date_dow_name given wrong day name
date.head()
```

```python
date.drop(columns=['product_id','city_id','orders']).head()  
```

```python
# is weekend 
date.drop(columns=['product_id','city_id','orders']).head()  
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
```

# Extract Quarter 

```python
date['quarter'] = date['date'].dt.quarter
date.drop(columns =['product_id','city_id','orders']).head()
```

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
date.drop(columns =['product_id','city_id','orders']).head()
```
