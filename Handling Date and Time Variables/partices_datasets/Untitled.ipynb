```python
sales_datasets = sales_datasets.drop(columns=['Dept','Weekly_Sales','IsHoliday'],axis = 1)
```

```python
# here our data we converting the data into the datatime datatype 
sales_datasets['Date'] = pd.to_datetime(sales_datasets['Date'],format='%d/%m/%Y')  # it has the differnt format in in data_time in the datafrme 
```

```python
sales_datasets.info()
```

```python
sales_datasets.head()
```

```python
# here we add the year values in the datafrme which is helpfull 
sales_datasets['date_years'] = sales_datasets['Date'].dt.year
sales_datasets.head()
```

```python
# here we extraing the month form the dataframe 
sales_datasets['date_month'] = sales_datasets['Date'].dt.month
sales_datasets.head()
```

```python
# here we extraing the days from the data to new columns 
sales_datasets['date_day'] = sales_datasets['Date'].dt.day
sales_datasets.head()
```

```python
# from the dataframe we extraing the data in month name 
sales_datasets['date_month_name'] = sales_datasets['Date'].dt.month_name()
sales_datasets.head()
```

```python
# here we extraing the week of day in the dataframe
sales_datasets['date_dow'] = sales_datasets['Date'].dt.dayofweek
sales_datasets.head()  # we haved extrated days of 
```

```python
# here we finding the data_dow name 
sales_datasets['date_dow_name'] = sales_datasets['Date'].dt.day_name()
sales_datasets.head()
```

```python
# here now we finding the weekind 
sales_datasets['date_is_weekend'] = np.where(sales_datasets['date_dow_name'].isin(['sunday','saturday']),1,0)
sales_datasets.head()
```

```python
# here we fidning the week days in the dataframe in whisch is help for the dataframe
# sales_datasets['weekend_days'] = sales_datasets['Date'].dt.isocalendar()
sales_datasets['date_week'] = sales_datasets['Date'].dt.isocalendar().week
sales_datasets.head()
# month ---> date_week we finding it 
```

```python
# now extrating the quators in the dataframe 
sales_datasets['queator'] = sales_datasets['Date'].dt.quarter
sales_datasets.head()
```

```python
# now we extraing the semester form the dataframe which is usefull 
sales_datasets['semester'] = np.where(sales_datasets['Date'].isin([1,2]),1,2)
sales_datasets.head()
# here form the date we extrating the datafame 
```
