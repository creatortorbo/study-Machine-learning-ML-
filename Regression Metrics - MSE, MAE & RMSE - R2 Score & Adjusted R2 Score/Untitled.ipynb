```python
from matplotlib import pyplot as plt 
import pandas as pd
import seaborn as sns
import numpy as  np
```

```python
df = pd.read_csv('placement.csv')
df.head()
```

```python
plt.figure(figsize =(14,6))
sns.scatterplot(x = 'cgpa',y = 'package',data = df ,color = 'g')
plt.show()
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
plt.figure(figsize =(14,6))
sns.scatterplot(x = 'cgpa',y = 'package',data = df ,color = 'g')
plt.plot(x_train,lr.predict(x_train),color = 'red')
plt.xlabel('CGPA')
plt.ylabel('Pakage (in lpa)')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred = lr.predict(x_test)
```

```python
y
```

```python
y_test.values
```

```python
print("MAE",mean_absolute_error(x_test,y_pred))
```

```python
print("MSE",mean_squared_error(x_test,y_pred))
```

```python
print("RMSE",np.sqrt(mean_squared_error(x_test,y_pred)))
```

```python
print("R2score",r2_score(x_test,y_pred))
```
