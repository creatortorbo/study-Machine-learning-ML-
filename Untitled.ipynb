```python
x =df.iloc[:,0:5]
```

```python
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size=0.2)
```

# Aam zindigi

```python
# adding simple imputer to fever col(here we removing the nan vlaues in  the dataframe)
si = SimpleImputer()
x_train_fever = si.fit_transform(x_train[['fever']])

# also the test data 
x_test_fever = si.fit_transform(x_test[['fever']])

x_train_fever.shape 


```

```python
x_train_fever.shape
```

```python
# Oridinal encoding --->cough 
oe = OrdinalEncoder(categories = [['Mild','Strong']])
x_train_cough = oe.fit_transform(x_train[['cough']])

# also the test data 
x_test_cough = oe.fit_transform(x_test[['cough']])

x_train_cough.shape
```

```python
# OneHotEncoding ---> gender,city 
ohe = OneHotEncoder(drop = 'first',sparse_output=False)
# here in the onehotencoding we are removing the one columns form the dataframe
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])

# also the test data 
x_test_gender_city = ohe.fit_transform(x_test[['gender','city']])

x_train_gender_city.shape
```

```python
# Extracting Age 
x_train_age = x_train.drop(columns = ['gender','fever','cough','city']).values

# also the test data 
x_test_age = x_test.drop(columns = ['gender','fever','cough','city']).values 

x_train_age.shape 
```

```python
x_train_transformed = np.concatenate((x_train_age,x_train_fever,x_train_gender_city,x_train_cough),axis = 1)
x_test_transformed = np.concatenate((x_test_age,x_test_fever,x_test_gender_city,x_test_cough),axis = 1)
```

```python
x_train_transformed_df = pd.DataFrame(x_train_transformed)
x_test_transformed_df = pd.DataFrame(x_test_transformed)
```

```python
# x_train_transformed_df # here we train data 
x_train_transformed_df
```

```python
x_test_transformed_df # here we testing data in the dataframe 
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr .fit(x_train_transformed_df, y_train)
```

```python
y_pred = lr.predict(x_test_transformed_df)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
lr = LogisticRegression()
from sklearn.model_selection import cross_val_score

np.mean(cross_val_score(lr,x_train_transformed_df,y,scoring='r2'))
```
