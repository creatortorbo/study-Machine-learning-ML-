```python
# creating the empty dataframe using 
df = pd.DataFrame()
```

```python
df # hrer empty dataframe
```

```python
for i in range(1,4):
      response = requests.get('https://mocki.io/v1/e3211270-ebe3-48f5-8265-df20465d7d53')
      df_new = pd.DataFrame(response.json()['ctRoot']).head()[['_id','name','dob','address','score','email','description','salary']]
      df = df.append(df_new,ignore_index = True)
```
