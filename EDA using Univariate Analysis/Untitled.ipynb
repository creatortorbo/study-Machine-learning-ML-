```python
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.shape
```

# Categorical Data
a.Countplot 

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Survived',data=df,palette = "muted",legend = False)
plt.show()
df['Survived'].value_counts()
# here using the colors code for the palette in the seaborn
```

```python
# hrer it is not working in this 
plt.figure(figsize = (5,5))
sns.countplot(df['Survived'].value_counts()).plot(kind ='bar')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Pclass',data=df,palette = "muted",legend = False)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Sex',data=df,palette = "muted")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Embarked',data=df,palette = "muted",legend = False)
plt.show() # Embarked ---> where the people come form(in ship)
```

# b.pieChart --> when we fom of %

```python
df['Sex'].value_counts().plot(kind="pie",autopct = '%.2f')
plt.show()
```

```python
# here using for percentage 
df['Embarked'].value_counts().plot(kind="pie",autopct = '%.2f')
plt.show()
```
