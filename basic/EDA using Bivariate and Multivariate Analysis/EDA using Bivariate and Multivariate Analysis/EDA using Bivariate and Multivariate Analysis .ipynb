```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt 
```

```python
# here importing the data from the seaborn 
tips = sns.load_dataset('tips')
tips.sample(5)
```

```python
# here come with csv file 
titanic = pd.read_csv('train.csv')
titanic.sample(6)
```

# 1.Scatterplot  (Numerical - Numerical)

# Bivariate anylyis

```python
# sns.scatterplot(tips['total_bill'],tips['tip'],hue = tips['sex'])
sns.scatterplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex')
plt.show()
# x -- is smoker and 0 ---> is not smoker 
```

# Multivariate analyisis

```python
sns.scatterplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',style='sex',size = 'size')
plt.show()
```

# Bivariate anylyis

# Box Plot (Numerical - Categorical)

```python
sns.boxplot(x ='Sex',y='Age',data=titanic,hue = 'Survived',palette = 'BuPu_r')
plt.show()
```

# 4.Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist =False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist =False)
plt.show()
```

# 5 HeatMap (categorical - Categorical)

```python
titanic.head()
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
# here we created the dataframe
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
plt.show()
```

# 6.ClusterMap(Categorical - Categorical)

```python
titanic.groupby('Pclass').mean()['Suevived']
```

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
plt.show()
```
