```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv',usecols = ['GarageQual','FireplaceQu','SalePrice'])
df.sample(5)
```

```python
# here we finding the % of data which have missing values
df.isnull().mean()*100
```

```python
df['GarageQual'].value_counts().plot.bar()
plt.show()
```

```python
df['GarageQual'].mode()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df[df['GarageQual'] =='TA']['SalePrice'].plot(kind = 'kde',ax = ax)

df[df['GarageQual'].isnull()]['SalePrice'].plot(kind = 'kde',ax = ax,color = 'red')

# Assuming you have already created your plot and obtained the handles and labels
ax = plt.gca()  # Get the current Axes instance
lines, labels = ax.get_legend_handles_labels()
labels = ['House with TA', 'Houses with NA']

# Set the title using ax.set_title()
ax.set_title('GarageQual')

# Optionally, you can add a legend using ax.legend()
ax.legend(lines, labels, loc='best')

plt.show()  # Display the plot
```

```python
temp = df[df['GarageQual']=='TA']['SalePrice']
```

```python
df['GarageQual'].fillna('TA',inplace =True)
# here we imputed the garagequal 
```

```python
df['GarageQual'].value_counts().plot(kind = 'bar')
plt.show ()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

temp.plot(kind = 'kde',ax = ax)

# distribution of the variable after imputaion 
df[df['GarageQual'] == 'TA']['SalePrice'].plot(kind = 'kde',ax = ax,color ='red')

lines, labels = ax.get_legend_handles_labels()
labels = ['Original variable','imputed variable']
ax.legend(lines,labels,loc ='best')

# add title 
plt.title('GarageQual')
plt.show()
```
