```python
from sklearn.datasets  import make_regression
from matplotlib import pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
x,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets = 1,noise = 80,random_state = 13)
```

```python
plt.scatter(x,y,color = 'g')
plt.show()
```

```python
# lets apply OLS
# here in the finding the slop and intercepter 
from sklearn.linear_model import LinearRegression 
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'r')
```

```python
# lets apply Gradinet Descent assuming slope is constant m = 78.35
# and lets assume the staring value for intercept b = 0
y_pred = ((78.35 * x) + 0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'OLS')
plt.plot(x,y_pred,color = '#00a65a',label= 'b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
# Iteration 1
# here we finding the slop 
loss_slope = - 2 * np.sum(y - m*x.ravel()- b)
loss_slope
```

```python
# lets take learning rate = 0.1
lr = 0.1

step_size = loss_slope*lr   # here when we mutipling the learning * slope 
step_size 
```

```python
# calculating the new intercpt
b = b - step_size 
b 
```

```python
y_pred1 = ((78.35 * x) + b).reshape(4)
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'OLS')
plt.plot(x,y_pred1,color = '#00a65a',label= 'b = {}'.format(b))
plt.plot(x,y_pred,color = '#A3E4D7',label= 'b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2
# here we finding the slop 
loss_slope = - 2 * np.sum(y - m*x.ravel()- b)
loss_slope
```

```python
step_size = loss_slope*lr   # here when we mutipling the learning * slope 
step_size 
```

```python
# calculating the new intercpt
b = b - step_size 
b 
```

```python
y_pred2 = ((78.35 * x) + b).reshape(4)
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'OLS')
plt.plot(x,y_pred2,color = '#00a65a',label= 'b = {}'.format(b))
plt.plot(x,y_pred1,color = '#00a65a',label= 'b = {}'.format(b))
plt.plot(x,y_pred,color = '#A3E4D7',label= 'b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 3
# here we finding the slop 
loss_slope = - 2 * np.sum(y - m*x.ravel()- b)
loss_slope
```

```python
step_size = loss_slope*lr   # here when we mutipling the learning * slope 
step_size 
```

```python
# calculating the new intercpt
b = b - step_size 
b 
```

```python
y_pred3 = ((78.35 * x) + b).reshape(4)

plt.figure(figsize=(14,6))
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'OLS')
plt.plot(x,y_pred3,color = '#00a65a',label= 'b = {}'.format(b))
plt.plot(x,y_pred2,color = '#00a65a',label= 'b = {}'.format(b))
plt.plot(x,y_pred1,color = '#00a65a',label= 'b = {}'.format(b))
plt.plot(x,y_pred,color = '#A3E4D7',label= 'b = 0')
plt.legend()
plt.show()
```

```python
# here it finded the best fited line in the given 
b = -100
m = 78.35
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*x.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * x + b

  plt.plot(x,y_pred)

plt.scatter(x,y)
plt.show()
```
